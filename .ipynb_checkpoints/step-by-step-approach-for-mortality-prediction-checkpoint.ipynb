```python
x = df.drop(columns='outcome')

y = df[['outcome']]
```

# Imputing Missing Values

## Impute the Float and Int values with mean values

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
col = x.select_dtypes(include='float64').columns
```

```python
col 
```

```python
si.fit(x[col])
```

```python
x[col] = si.transform(x[col])
```

```python
x.info()
```

# Impute the dependent Variable with most frequent value

```python
SI =  SimpleImputer(missing_values=np.nan, strategy="most_frequent")
```

```python
SI.fit(y)
```

```python
y = SI.transform(y)
```

```python
y = pd.DataFrame(y, columns=['outcome'], dtype='int64')
```

```python
y
```

# ETL

```python
df_final = x.copy()

df_final['outcome'] = y
```

### As mentioned earlier, this is an highly imbalanced dataset out of all patients only 14% patients were died but remianing 86% of patients were alive , due to this imbalance in nature less bias towards major class (i.e. 0 alive) and high variance on test data due to this reason model has predicted all patients as alive, moreover as observed from classification report the accuracy is 85% but f1 score for Minority class is 36%, to deal this problem we have to balance the data between majority class (i.e. 0) and Minority calss (i.e. 1), there were different techniques available but for this study I will be using **SMOTE ('Synthetic Minority Oversampling Technique')**

# SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
X = x.drop(columns='ID')
Y = y.copy()
```

```python
smote = SMOTE(sampling_strategy='minority')
```

```python
x_s, y_s = smote.fit_resample(X, Y)
```

### Train Test Split 

1.   70% training data
2.   30% test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_s, y_s, test_size=0.3, random_state=123)
```

## Model on Balanced Dataset

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(random_state=123, seed=123)
```

```python
xgb.fit(X_train, Y_train)
```

```python
pred_b = xgb.predict(X_test)
```

### Compariosn with actual values

```python
con = confusion_matrix(Y_test, pred_b)

con
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(con, annot=True, cmap='viridis', fmt='.0f');
plt.xlabel('Predicted', fontsize=14, weight='bold')
plt.ylabel('Actual', fontsize=14, weight='bold')
```

```python
tn, fp, fn, tp = confusion_matrix(Y_test, pred_b).ravel()

print(tn)
print(fp)
print(fn)
print(tp)
```

```python
print(classification_report(Y_test, pred_b))
```

##### Previous classification report (i.e. Imbalance data)

```python
print(classification_report(y_test, pred))
```
